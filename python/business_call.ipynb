```python
# demo loop for yelp calls
test_city = 'philadelphia'

r = requests.get(yelp_base + f'location={test_city}&price=3,4&limit=50&offset=0', headers=headers)
page_one = json.loads(r.content)

if int(page_one['total']/50) > 1:
    pages = int(page_one['total']/50)
    results = []
    results.append(page_one)
    for page in range(1, pages + 1):
        r = requests.get(yelp_base + f'location={test_city}&price=3,4&limit=50&offset={50*page}', headers=headers)
        additional_page = json.loads(r.content)
        results.append(additional_page)
```

```python
# loop through each business and convert to a city based json
city_json = {}

for i in range(len(results)):
    for j in range(len(results[i]['businesses'])):
        restaurant_name = results[i]['businesses'][j]['name']
        business_json = {}
        business_json['restaurant_name'] = restaurant_name 
        business_json['restaurant_location'] = [results[i]['businesses'][j]['coordinates']['latitude'],
                               results[i]['businesses'][j]['coordinates']['longitude']]
        business_json['restaurant_rating'] = results[i]['businesses'][j]['rating']
        business_json['restaurant_popularity'] = results[i]['businesses'][j]['review_count']
        business_json['restaurant_url'] = results[i]['businesses'][j]['url']
        business_json['restaurant_address'] = results[i]['businesses'][j]['location']['display_address'][0]
        business_json['restuarant_phone'] = results[i]['businesses'][j]['display_phone'] 
        business_json['restaurant_tags'] = []
        for k in range(len(results[i]['businesses'][j]['categories'])):
            business_json['restaurant_tags'].append(results[i]['businesses'][j]['categories'][k]['title'])
        city_json[restaurant_name] = business_json 
city_json
```

```python
# Import dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
from splinter import Browser 
import time
import pymongo
```

```python
# Setup Chromedriver
executable_path = {'executable_path': 'chromedriver/chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Scrape wiki site for top 5 cities in each state and territory in the us
url = 'https://en.wikipedia.org/wiki/List_of_largest_cities_of_U.S._states_and_territories_by_population'
```

```python
browser.visit(url)
```

```python
# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
# Use pandas to read for tables
tables = pd.read_html(url)
largest_cities = tables[0]
largest_cities
```

```python
# Store States in a list
states = []

for i in range(len(largest_cities['State, federaldistrict, orterritory'])):
    states.append(largest_cities['State, federaldistrict, orterritory'][i])
    
states
```

```python
# Store most populous city in a list
populous_lvl_1 = []
init_td = ''
polished_td = ''

for i in range(len(largest_cities['Most populous'])):
    init_td = largest_cities['Most populous'][i]
    #Remove any numbers from the string
    polished_td = ''.join([i for i in init_td if not i.isdigit()]) 
    # Append to most populous list
    populous_lvl_1.append(polished_td)
    
populous_lvl_1
```

```python
# Demo remove any characters not related to the city
sep = '('
stripped = largest_cities['2nd most populous'][0].split(sep, 1)[0]
stripped.rstrip()
```

```python
# store 2nd most populous city in a list
populous_lvl_2 = []
init_td = ''
stripped_td = ''
polished_td = ''
sep = '('

for i in range(len(largest_cities['2nd most populous'])):
    try:
        init_td = largest_cities['2nd most populous'][i]
        stripped_td = init_td.split(sep, 1)[0].rstrip()
        polished_td = ''.join([i for i in stripped_td if not i.isdigit()])
    except (AttributeError):
        polished_td = None
    populous_lvl_2.append(polished_td)  

populous_lvl_2
```

```python
# form new dataframe
locations_df = pd.DataFrame(states, columns=['State'])
locations_df['city_1'] = populous_lvl_1
locations_df['city_2'] = populous_lvl_2
locations_df['city_3'] = populous_lvl_3
locations_df['city_4'] = populous_lvl_4
locations_df['city_5'] = populous_lvl_5
locations_df
```

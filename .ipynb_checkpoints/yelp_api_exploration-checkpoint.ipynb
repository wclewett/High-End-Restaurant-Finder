```python
# import dependencies
import requests
import pandas as pd
import json
```

```python
yelp_base = 'https://api.yelp.com/v3/businesses/search?'
yellow = {'Authorization': 'bearer insert key here'}
# Dummy Call to retrieve total amount
r = requests.get(yelp_base + f'location=New York City&price=3,4&limit=50&offset=950', headers=yellow)
response_json = json.loads(r.content)
```
